```python
import pandas as pd
```

```python
df = pd.read_csv("vgsales.csv")
```

```python
df.head(5)
```

```python
import numpy as np
```

```python
def consoleCountry(row):
    if row["Platform"] == "X360":
        return "International"
    elif row["Platform"] == "3DO":
        return "International"
    elif row["Platform"] == "2600":
        return "International"
    elif row["Platform"] == "PC":
        return "International"
    elif row["Platform"] == "XOne":
        return "International"
    elif row["Platform"] == "XB":
        return "International"
    else:
        return "Japan"
df["Region"] = df.apply(lambda row: consoleCountry(row), axis=1)
df.head(5)
```

```python
import plotly.express as px
fig = px.histogram(df, x="Platform")
fig.show()
```

```python
fig = px.histogram(df, x="Region")
fig.show()
```
